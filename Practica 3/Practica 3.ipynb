```python
import pandas as pd
```

```python
df = pd.read_csv("farm_production_dataset.csv")
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
df.groupby('REF_DATE')['Production (metric tonnes)'].sum().plot()
plt.title("Grafica de produccion anual")
plt.ylabel("Produccion (toneladas metricas)")
plt.xlabel("AÃ±o")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
df.groupby('REF_DATE')['Total farm value (dollars)'].sum().plot()
plt.title("Grafica de valor de total de granjas por aÃ±o")
plt.ylabel("Valor total de las granjas")
plt.xlabel("AÃ±o")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
df.groupby('REF_DATE')['Average farm price (dollars per tonne)'].sum().plot()
plt.title("Grafica de valor de promedio de granjas por aÃ±o")
plt.ylabel("Valor promedio de granjas")
plt.xlabel("AÃ±o")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
df.groupby('GEO')['Seeded area (hectares)'].sum().plot(kind='bar')
plt.title("Grafica de area plantada por region")
plt.ylabel("Area plantada")
plt.xlabel("Region")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
df.groupby('GEO')['Production (metric tonnes)'].sum().plot(kind='bar')
plt.title("Grafica de produccion por region")
plt.ylabel("Produccion")
plt.xlabel("Region")
plt.show()
```
