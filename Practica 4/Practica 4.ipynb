```python
import pandas as pd
```

```python
df = pd.read_csv("farm_production_dataset.csv")
```

```python
df
```

```python
# @title Prueba de kruskal en el precio promedio de granja y su produccion

from scipy import stats

y = df['Average farm price (dollars per tonne)']
x = df['Production (metric tonnes)']

Kruskal = stats.kruskal(x,y, nan_policy='omit')
Kruskal
```

```python
# @title Prueba de kruskal en el precio promedio de granja y su rendimiento promedio

from scipy import stats

y = df['Average farm price (dollars per tonne)']
x = df['Average yield (kilograms per hectare)']

Kruskal = stats.kruskal(x,y, nan_policy='omit')
Kruskal
```

```python
# @title Prueba de kruskal en el precio promedio de granja y area plantada

from scipy import stats

y = df['Average farm price (dollars per tonne)']
x = df['Seeded area (hectares)']

Kruskal = stats.kruskal(x,y, nan_policy='omit')
Kruskal
```

```python
# @title Prueba de kruskal en el precio promedio de granja y su valor total

from scipy import stats

y = df['Average farm price (dollars per tonne)']
x = df['Total farm value (dollars)']

Kruskal = stats.kruskal(x,y, nan_policy='omit')
Kruskal
```
