```python
import pandas as pd
```

```python
df = pd.read_csv("farm_production_dataset.csv")
```

```python
# @title Datos estadisticos
df.describe(include='all')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
df.groupby('REF_DATE')['Production (metric tonnes)'].sum().plot()
plt.title("Grafica de produccion anual")
plt.ylabel("Produccion (toneladas metricas)")
plt.xlabel("AÃ±o")
plt.show()
```

```python
# @title Area plantada con respecto a la region
average_farm_size = df.groupby('GEO')['Seeded area (hectares)'].mean()
average_farm_size
```
