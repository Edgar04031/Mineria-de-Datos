```python
import pandas as pd
```

```python
df = pd.read_csv("farm_production_dataset.csv")
```

```python
df
```

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train = df[['total_farm_value','avg_farm_price', 'avg_yield', 'production']]
Y_train = df['ref_date']
```

```python
Regresion = linear_model.LinearRegression()
Regresion.fit(X_train, Y_train)
y_pred = Regresion.predict(X_train)
print('Coefficients: \n', Regresion.coef_)
print('Independent term: \n', Regresion.intercept_)
print("Mean squared error: %.2f" % mean_squared_error(Y_train, y_pred))
print('Variance score: %.2f' % r2_score(Y_train, y_pred))

plt.scatter(X_train.iloc[:,0], Y_train,  color='black')
plt.plot(X_train.iloc[:,0], y_pred, color='blue', linewidth=3)
```
